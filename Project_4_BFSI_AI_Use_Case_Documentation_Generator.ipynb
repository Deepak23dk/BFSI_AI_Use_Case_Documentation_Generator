# BFSI AI Use Case Documentation Generator

## Project Objective

This project generates realistic BFSI AI use cases using an Agentic AI workflow.

The system contains:

1. Planner Agent
2. Research Agent
3. Use Case Generator Agent
4. Risk Analysis Agent
5. Reviewer Agent
6. Score Agent
7. PDF Export Agent

Goal:
Generate professional BFSI AI documentation reports automatically.

## Agent Architecture

User Input
 ↓
Planner Agent
 ↓
Research Agent
 ↓
Use Case Generator Agent
 ↓
Reviewer Agent
 ↓
Score Agent
 ↓
PDF Export Agent
 ↓
Final Report

 STEP 2: AgenticAI_API Key Setup

```python
import google.generativeai as genai

API_KEY = input("Enter Your Gemini API Key: ")

genai.configure(api_key=API_KEY)

model = genai.GenerativeModel("gemini-2.5-flash")

print("API Connected Successfully")
```

STEP 2: Testing the Setup

```python
response = model.generate_content("Say hello")

print(response.text)
```

STEP 3: Create Planner Agent

```python
def planner_agent(sector, technique):

    return f"""
    Create a BFSI AI documentation plan.

    Sector: {sector}

    Technique: {technique}

    Generate:
    - Industry Context
    - AI Opportunities
    - Business Goals
    """
```

STEP 4: Create Research Agent

```python
def research_agent(plan):

    prompt = f"""
    You are a BFSI industry expert.

    Plan:
    {plan}

    Provide:

    1. Current Industry Trends
    2. AI Opportunities
    3. Business Challenges
    4. Future AI Scope
    """

    response = model.generate_content(prompt)

    return response.text
```

STEP 5: Create Use Case Generator Agent

```python
def usecase_generator_agent(plan):

    prompt = f"""
    {plan}

    Generate:

    1. Five BFSI AI Use Cases

    For each:
    - Problem
    - Technique
    - Workflow
    - Example
    - Impact
    - Risk

    After the report generate:

    REVIEW SECTION
    - Strengths
    - Weaknesses
    - Suggestions

    SCORE SECTION
    - Completeness Score /100
    - Professionalism Score /100
    - Overall Score /100
    """

    return model.generate_content(prompt).text
```

STEP 6: Create Reviewer Agent

```python
def reviewer_agent(report):

    prompt = f"""
    Review this BFSI report.

    Check:

    - Completeness
    - Clarity
    - Industry Relevance
    - Professional Quality

    Give detailed feedback.

    Report:

    {report}
    """

    response = model.generate_content(prompt)

    return response.text
```

STEP 7: Create Score Agent

```python
def score_agent(report):

    prompt = f"""
    Score this report.

    Give:

    Completeness Score /100
    Professionalism Score /100
    Relevance Score /100

    Final Overall Score /100

    Report:

    {report}
    """

    response = model.generate_content(prompt)

    return response.text
```

STEP 8: Create Dashboard Agent

```python
def dashboard_agent(
    sector,
    technique,
    score
):

    print("="*50)

    print("PROJECT DASHBOARD")

    print("="*50)

    print("Sector:", sector)

    print("Technique:", technique)

    print("\nExecution Status: SUCCESS")

    print("\nAI Evaluation:")

    print(score)

    print("="*50)
```

STEP 9: Create PDF Export Agent

```python
from fpdf import FPDF

def save_pdf(content):

    pdf = FPDF()

    pdf.add_page()

    pdf.set_font("Arial", size=12)

    content = content.encode(
        "latin-1",
        "replace"
    ).decode("latin-1")

    pdf.multi_cell(
        0,
        10,
        content
    )

    pdf.output(
        "BFSI_Report.pdf"
    )

    print("PDF Saved Successfully")
```

STEP 10: Main Execution Cell

```python
sector = input("Enter BFSI Sector: ")

technique = input("Enter AI Technique: ")

plan = planner_agent(
    sector,
    technique
)

report = usecase_generator_agent(
    plan
)

print(report)

dashboard_agent(
    sector,
    technique,
    "Generated Successfully"
)

save_pdf(report)
```
